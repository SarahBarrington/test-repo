```python
import matplotlib as mpl
#mpl.rcParams['axes.color_cycle'] = ['#7FB5C7', '#E63E65', '#5B5BC9', '#55D957']

def adjust_tree_figure(fname):
    import re
    with open(fname, 'r') as dotfile:
        newdotstr = re.sub(r'mse = \d+.\d+..', '', dotfile.read())
        newdotstr = re.sub(r'value', 'bikes', newdotstr)
        newdotstr = re.sub(r'shape="box"', 
                           'shape="box", style="rounded", fontname="verdana", fontsize=8', 
                           newdotstr)
        with open(fname, 'wb') as newdotfile:
            newdotfile.write(newdotstr)

%matplotlib inline
```

```python
import numpy as np
np.random.seed(1)
```

*** Loading and visualising the data ***

```python
import pandas as pd
bikes = pd.read_csv('bikes.csv')
bikes.head() # Use pd.read to get a NICE table, not just a read csv file!
bikes #added this in to see whole thing
bikes.shape # basically the STR in R but not as good lol 
```

```python
bikes.temperature #gets you the temperature column of bikes
bikes.temperature * 2 # times column values by 2
(bikes.temperature + bikes.humidity).sum()
```

```python
from matplotlib import pyplot as plt 
# plot number of bikes vs the temperature for a given day
plt.figure(figsize=(8,6)) #actual figure size in inches!
plt.plot(bikes['temperature'], bikes['count'], 'o') #temp x axis, count y axis
plt.xlabel('temperature')
plt.ylabel('bikes')
plt.show()
```

*** Training the first regressor ***

```python
from sklearn.tree import DecisionTreeRegressor #using the inbuilt functionality from sklearn package
import numpy as np

regressor = DecisionTreeRegressor(max_depth=2)
regressor.fit(np.array([bikes['temperature']]).T, bikes['count']) #temp as input, gives count back as output. Output is just a description of the model but it gives you all the parameters. 
```

```python
regressor.predict(5.) #for 5 degrees C it will predict the bikes for you!
```

```python
regressor.predict(np.array([[20.,5.]]).T)
```

```python
xx = np.array([np.linspace(-5, 40, 100)]).T ##xx is just an array covering all the temperatures, equally spaced values- prediction array

plt.figure(figsize=(8,6))
plt.plot(bikes['temperature'], bikes['count'], 'o', label='observation') #is plotting all the predictions and thus showing you the model pattern it's picked up-visualised below 
plt.plot(xx, regressor.predict(xx), linewidth=4, alpha=.7, label='prediction')
plt.xlabel('temperature')
plt.ylabel('bikes')
plt.legend()
plt.show()
#note that trees can only predict values they've seen before e.g. cant predict outside the range of original values
```

```python
from sklearn.tree import export_graphviz  #visualising the rules we have learned from the data. Takes the tree as input and converts it into a figure 
export_graphviz(regressor, out_file='tree.dot', feature_names=['temperature']) #'regressor' is the input 
adjust_tree_figure('tree.dot')
```

*** Multiple regression ***

```python
regressor = DecisionTreeRegressor(max_depth=2) #can set the tree depth to be higher for more divison in feature space but takes v long time to compute 
regressor.fit(bikes[['temperature', 'humidity']], bikes['count']) # so now we've added humidity in
```

```python
export_graphviz(regressor, out_file='tree_multi.dot', feature_names=['temperature','humidity'])
adjust_tree_figure('tree_multi.dot')
```

```python
!dot -Tpng tree_multi.dot > tree_multi.png # to convert the tree in a png image
```

```python
nx = 30
ny = 30
# creating a grid of points
x_temperature = np.linspace(-5, 40, nx)
y_humidity = np.linspace(20, 80, ny)
xx, yy = np.meshgrid(x_temperature, y_humidity)
# evaluating the regresson on all the points
z_bikes = regressor.predict(np.array([xx.flatten(), yy.flatten()]).T)
zz = np.reshape(z_bikes, (nx, ny))


fig = plt.figure(figsize=(8, 8))
# plotting the predictions
plt.pcolormesh(x_temperature, y_humidity, zz, cmap=plt.cm.YlOrRd) # Plotting in 3D by using dot size for number of bikes.
plt.colorbar(label='bikes predicted') # add a colorbar on the right. The colours show the predicted numbers (partitioning of feature space by tree shown)
# plotting also the observations
plt.scatter(bikes['temperature'], bikes['humidity'], s=bikes['count']/25.0, c='g')
# setting the limit for each axis
plt.xlim(np.min(x_temperature), np.max(x_temperature))
plt.ylim(np.min(y_humidity), np.max(y_humidity))
plt.xlabel('temperature')
plt.ylabel('humidity')
plt.show()

plt.show()
```

*** Validating our model ***

```python
# comparing our models 
from sklearn.metrics import mean_absolute_error # MAE is the average of (lots of differences between 2 numbers, e.g. true and predicted)
mean_absolute_error(bikes['count'], regressor.predict(bikes[['temperature','humidity']]))
```

```python
from sklearn.cross_validation import cross_val_score
scores = -cross_val_score(regressor, bikes[['temperature', 'humidity']], 
                          bikes['count'], scoring='mean_absolute_error', cv=10) #10 splits
```

```python
scores #gives us a value at each score
```

```python
scores.mean() #mean absolute error
```

```python
regressor2 = DecisionTreeRegressor(max_depth=100)
regressor2.fit(bikes[['temperature', 'humidity']], bikes['count']) #introduce second model to compare
```

```python
mean_absolute_error(bikes['count'], regressor2.predict(bikes[['temperature','humidity']]))
```

```python
scores = -cross_val_score(regressor2, bikes[['temperature', 'humidity']], 
                          bikes['count'], scoring='mean_absolute_error', cv=10)
scores.mean() #higher MAE. This is overfitting. Nightmare for DS! Cross validation very important. Think of it as a student who just learns all the questions without generalising over the whole examwith a bit of intuition!
# Mean absolute error isnt enough, it needs a cross validation. Lower CV score is better
```

```python
# feature importance
plt.bar([0,1], regressor2.feature_importances_, color ='#7FB5C7') #represents temp & humidity, just need labels
plt.xticks([0.3,1.3], ['temperature', 'humidity'])
```

*** From a tree to a forest ***

```python
from sklearn.ensemble import RandomForestRegressor #deciison tree = high var model. E.g. sensitive to changes in the sample. So use forest to minimise this variation- run many times and average the variation at the end
# bootstrapping (subsamplee with replacement) is used to grow many trees and test. 
# boosting can also be used. Assigns weights to samples. Do a regressor, get an error, then try and reweight and repeat to see if there is an imrovement. Carry on until model has reached optimal
# trade off between the 2- boosting v accurate but BS much faster. Accuracy vs training time. 
# ensemble can be applied to any sort of regression

forest = RandomForestRegressor(max_depth=3, n_estimators=100) #100 trees
forest.fit(np.array([bikes['temperature']]).T, bikes['count']) #fit will do the training for ya. Can use for any regressor in sklearn (both fit & predict)
# Allows plug & play with regressor e.g. compare single tree vs random forest. 

# Rand forest is just a more powerful/accurate/expensive decision tree approach
```

```python
xx = np.array([np.linspace(-5, 40, 100)]).T

plt.figure(figsize=(8,6))
plt.plot(bikes['temperature'], bikes['count'], 'o', label='observation')
plt.plot(xx, forest.predict(xx), linewidth=4, alpha=.7, label='prediction')
plt.xlabel('temperature')
plt.ylabel('bikes')
plt.legend()
plt.show()

# averaged step functions so doesnt look as step-function like as the straight up decision tree
# note bump in centre- due to outliers. USeful e.g. frauds in bank accounts! Sometimes just data faults & want to remove...
```

```python
forest.estimators_[1] # gives the array of each forest parameters, e.g. this will give you the first tree 
```

*** Prediction intervals ***

```python
def predict_interval(x, ensamble, z=1.96): #z = critical value, determines the length of interval
    """
    Input
     x, features in input
     ensamble, instance of an ensamble regressor
     z, critical value. Determines the confidence level:
     
           Level  z
            75%  1.15
            90%  1.64
            95%  1.96 #just 95% confidence level, eg. 95% of values are in that range.
            99%  2.58
     
    Output
     lower bound, center and high bound of the confidence interval #however assumes a normal distribution, which is usually OK in linear regression but perhaps not always here
    """
    predictions = [estimator.predict(x) for estimator in ensamble.estimators_]
    mean = np.mean(predictions, 0)
    std = np.std(predictions, 0)
    return mean - z*std, mean, mean + z*std
```

```python
plt.figure(figsize=(8,6))
plt.plot(bikes['temperature'], bikes['count'], 'o', label='observation', zorder=0)
low, mean, high = predict_interval(xx, forest)
plt.plot(xx, mean, linewidth=4, alpha=.7, label='prediction')
plt.fill_between(xx.T[0], low, high, facecolor='blue', alpha=0.5, label='Prediction interval')
plt.xlabel('temperature')
plt.ylabel('bikes')
plt.legend(loc='lower right')
plt.show()

#tells you how confident you can be in your estimation. E.g. at 15 degrees we see there is a big interval- so what went on there? Something fishy going on... 
# If i predict 20, there is alower and upper boundary of bikes to stock to feel safe. 
# good to tell the customer how confident you are in prediction. Quantifies risk. 
```
